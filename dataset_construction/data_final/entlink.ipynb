```python
import pandas as pd
import json
```

```python
ent_map = json.load(open('./reversed_dict_second.json'))
```

```python
ent_map
```

```python
len(ent_map)
```

```python
ent_map_keys = list(ent_map.keys())
```

```python
for k in ent_map_keys:
    if '\n' in k:
        print(k)
```

```python
ent_map_cleaned = {}
for k, v in ent_map.items():
    ent_map_cleaned[k.strip(' \n')] = v.strip(' \n')
```

```python
len(ent_map_cleaned)
```

```python
json.dump(ent_map_cleaned, open('./reversed_dict_second_cleaned.json', 'w'), indent=4)
```

```python
vicuna_entlink = pd.read_csv('./EGIRIS_tfdf.csv', sep='\t', dtype='string')
```

```python
vicuna_entlink
```

```python
print(len(pd.concat([vicuna_entlink['Subject'], vicuna_entlink['Object']]).unique()))
print(len(vicuna_entlink['Md5'].unique()))
```

```python
vicuna_entlink['Subject'] = [x.strip(' \n') for x in vicuna_entlink['Subject']]
vicuna_entlink['Object'] = [x.strip(' \n') for x in vicuna_entlink['Object']]
```

```python
print(len(pd.concat([vicuna_entlink['Subject'], vicuna_entlink['Object']]).unique()))
print(len(vicuna_entlink['Md5'].unique()))
```

```python
vicuna_entlink = vicuna_entlink[vicuna_entlink['Subject'].isin(ent_map_cleaned)]
vicuna_entlink = vicuna_entlink[vicuna_entlink['Object'].isin(ent_map_cleaned)]
```

```python
print(len(pd.concat([vicuna_entlink['Subject'], vicuna_entlink['Object']]).unique()))
print(len(vicuna_entlink['Md5'].unique()))
```

```python
vicuna_entlink['Subject'] = [ent_map_cleaned[x] for x in vicuna_entlink['Subject']]
vicuna_entlink['Object'] = [ent_map_cleaned[x] for x in vicuna_entlink['Object']]
```
